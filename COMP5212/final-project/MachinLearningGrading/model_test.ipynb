```python
train = TabularDataset("../data/train.csv")
test = TabularDataset("../data/validation.csv")
label = 'Label'
#predictor = TabularPredictor.load("../MachinLearningGrading/AutogluonModels/ag-20240417_070035")
predictor = TabularPredictor(label=label,eval_metric='f1',).fit(train,num_gpus=1,auto_stack=True,presets='best_quality')
```
